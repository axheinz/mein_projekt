```python
for id in range(0,zeilen):
    einzel_df = pd.json_normalize(data['resultTables'][0]['rows'][id]).T
    einzel_df.columns = einzel_df.iloc[0]
    einzel_df = einzel_df[1:]
    df_liste.append(einzel_df)
```

```python
df_komplett = pd.concat (df_liste)
```
